```python
%matplotlib inline
import matplotlib.pyplot as plt
#This is a timeseries
fig = plt.figure()
ax=fig.add_subplot(1,1,1)
ax.set_ylabel('flood', color='r')
data['flood']['Pearl'].plot(ax=ax, label='Pearl')
```

```python
#combine multiple timeseries
fig = plt.figure()
ax=fig.add_subplot(1,1,1)
ax.set_ylabel('flood', color='r')
data['flood']['Pearl'].plot(ax=ax, label='Pearl')
data['flood']['Yellow'].plot(ax=ax, label='Yellow')
data['flood']['Yangtze'].plot(ax=ax, label='Yangtze')
ax.legend(ncol=3)
```

```python
#plot data of different types/on different scales
fig = plt.figure()
ax1=fig.add_subplot(1,1,1)
ax2 = ax1.twinx()
ax1.set_ylabel('flood', color='r')
data['flood']['Pearl'].plot(ax=ax1, color='r')
ax2.set_ylabel('precip', color='b')
data['precip']['Pearl'].plot(ax=ax2, color='b')
```

```python
colors = {1:'SteelBlue', #January
         2:'LightSkyBlue', #February
         3:'GreenYellow', #March
         4:'ForestGreen', #April
         5:'OliveDrab', #May
         6:'Pink', #June
         7:'DeepPink', #July
         8:'MediumVioletRed',#August
         9:'Gold', #September
         10:'Orange',#October, 
         11:'OrangeRed', #November
         12: 'MidnightBlue', #December
         }
```

```python
#Color each month differently

pf = data['flood']['Pearl']
pi = data['inundation']['Pearl']
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
for m in range(1,13):
    ax.scatter(pf[pf.index.month==m],
               pi[pi.index.month==m], 
               c=colors[m])
```
