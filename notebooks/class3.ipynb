```python
#Contextualizing it?
pearl = data['flood']['Pearl']
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
pearl.hist(ax=ax)
ax.axvline(pearl.mean(), label="mean", color='r')
ax.axvline(pearl.median(), label="median", color='m')
std = pearl.std()
ymin, ymax = ax.get_ylim()
ax.vlines([-std, std], ymin, ymax, label="std", color='g')
ax.legend()
```
