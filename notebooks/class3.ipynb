```python
#Is there a quick way to find the correlation between all 5 variables?
#first package the data into an array
varlist = ['discharge', 'flood', 'inundation', 'ndvi', 'precip']
#make a dataframe for packaging the variables of your data
#list comprehension
darr = np.vstack((data[v]['Pearl'].values for v in varlist))
#this arranges the data (variable X observation)
#look at the documentation to see how the data should be organized
#http://docs.scipy.org/doc/numpy-1.6.0/reference/generated/numpy.corrcoef.html
cmat = np.corrcoef(darr)
print cmat
```

```python
#Lets filter out nan
#throw out any day where at least 1 variable is missing
clean = []
#transpose the data to (observation X variable) 
#because lost comprehensions go over rows
for day in darr.T: 
    if np.isfinite(day).all():
        clean.append(day)
#turn into an array
clean = np.vstack(clean)
#clean is now (obs x var) 
#turn it into (var x obs) for corrcoef
corr = np.corrcoef(clean.T)
print corr
```

```python
#simplify?
#find all the columns where all the elements are finite
#axis=0: true if all the rows of the column have finite elements
mask = np.isfinite(darr).all(axis=0)
#sanity check that they match up
print darr.shape, mask.shape
#for all rows, choose the columns returned by mask
corr = np.corrcoef(darr[:, mask])
print corr
```

```python
#visualize this?
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
im = ax.matshow(corr)
fig.colorbar(im, ax=ax)
ax.set_xticks(range(len(varlist)))
ax.set_yticks(range(len(varlist)))
ax.set_xticklabels(varlist)
ax.set_yticklabels(varlist)
```
