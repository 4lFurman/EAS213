```python
#make the scatter colors mean something
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
#change the bounding box for the river delta
m = Basemap(projection='cyl', ax=ax, resolution = 'h',
            llcrnrlat=10,urcrnrlat=60, llcrnrlon=80,urcrnrlon=140)
border_color = '0.8'
m.drawcoastlines(color=border_color)
m.drawcountries(color=border_color)
m.drawmapboundary(color=border_color)
m.drawrivers(color='lightblue', linewidth=.5)
ax.set_title("Maximum Flood Severity")
deltas = [('Pearl', 22.5333, 113.7333), 
          ('Yellow',34.9219, 97.5119), 
          ('Yangtze',31.151636,121.8083)]

xs = []
ys = []
vals = []
for dn, lat, lon in deltas:
    #lats = y axis
    #lons = x axis
    x, y = m(lon, lat)
    val = data['flood'][dn].max()
    ax.text(x+2, y, dn)
    xs.append(x)
    ys.append(y)
    vals.append(val)
#passing in a list of values for colors
#colors the scatter based on the data value
im = m.scatter(xs,ys, s=50,c=vals, zorder=100) 
m.colorbar(im, ax=ax)
```
