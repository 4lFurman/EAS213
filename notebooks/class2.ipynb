```python
#We can also plot specific ranges, like the first and last year
pf = data['flood']['Pearl']
fig = plt.figure()

ax1 = fig.add_subplot(3,1,1)
ax1.set_title('1990')
pf[pf.index.year==1990].plot(ax=ax1)

ax2.set_title('2013')
ax2 = fig.add_subplot(3,1,2)
pf[pf.index.year==2013].plot(ax=ax2)

ax3.set_title('Every 10 Days')
ax3 = fig.add_subplot(3,1,3)
pf[::10].plot(ax=ax3)
```
