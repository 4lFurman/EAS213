Delta Flood Risk Factor Analysis
===========================

Variables
==================
0. Flood Severity:
    * Severity is a 1, 1.5, 2 scale 
    * magnitude=log(duration * severity * area)

1. Precipitation:
    * daily accumulated rainfall (mm/day)
2. Inundation
    * % of the land containing water
3. NDVI
    * Normalized Difference Vegetation Index (-1.0 to 1.0)
4. Discharge
    * m^3/s flow of water

```python
#Lets take a closer look at floods
#To get a section is called slicing, lets do the first year
#slices are [start:stop:step] with any bit being optional
print "full slice: "
print dm['1990-01-01':'1990-12-31']
#this is equivalent to
print "implicit start: "
print dm[:'1990-12-31']
#and can be shortened further to
print "fancy slice: "
print dm[dm.index.year==1990]
```

```python
#now lets do the last year
#slices are [start:stop:step] with any bit being optional
print "full slice: "
print dm['2013-01-01':'2013-08-31']
#this is equivalent to
print "implicit start: "
print dm['2013-01-01':]
#and can be shortened further to
print "fancy slice: "
print dm[dm.index.year==2013]
```

```python
#And lets grab every 10th day
#slices are [start:stop:step] with any bit being optional
print "full slice: " 
print dm['1990-01-01':'2013-08-31':10]
#this is equivalent to
print "implicit start: "
dm[::10]
#and can be shortened further to
print "fancy slice: " 
dm[dm.index.day%10==1]
```

```python
#Lets compute some statistics with these slices:
dmn = dm[dm.index.year==1990].min()
dmx = dm[dm.index.year==1990].max()
#fancy indexing can be nested
dmd = (dm[dm.index.year==1990]>0).sum()
print "Beginning: min = ", dmn, "max =", dmx,"days = ", dmd
```

```python
#To simplify further, assign the shortened array to a variable
dme = dm[dm.index.year==2013]
dmn = dme.min()
dmx = dme.max()
dmd = (dme>0).sum()
print "Beginning: min = ", dmn, "max =", dmx,"days = ", dmd
```

```python
#To simplify further, assign the shortened array to a variable
dme = dm[::10]
dmn = dme.min()
dmx = dme.max()
dmd = (dme>0).sum()
#simplify the string using string formatting
print "Beginning: min = {} max = {}, days = {}".format(dmn,dmx,dmd)
```
