Measuring Temperature
------------------------

1 Kelvin = 1 °Celsius = 9/5 °Fahrenheit

### From Kelvins:
* [°C] = [K] − 273.15
* [°F] = [K] × 9⁄5 − 459.67

### To Kelvins:
* [K] = [°C] + 273.15
* [K] = ([°F] + 459.67) × 5⁄9


```python
#Let's convert 100 degrees Kelvin to Celsius:
100-273.15
```

```python
#Now write a function kelvin2celsius to convert from Kelvins to Celsius

def kelvin2celsius(K):
    """Converts Kelvin Temperature to Celsius Temperature
    
        Parameters
        ----------
        K: double
            Kelvin Temperature
        
        Returns
        -------
        :double
            Celsius Temperature
    """
    return K-273.15
```

```python
#And test the functions:
print kelvin2celsius(100)
#Why do you think the numbers differ? Is the function implemented correctly?
```

```python
#lets make a list of Kelvin temperatures
ktemps = range(0,100,10)
#and now, lets make a lookup table of our kelvin temperatures:
k2c = dict() #empty dictionary

for k in ktemps: #loop over temperatures
    k2c[k] = kelvin2celsius(k) #store result of function in dictionary

print k2c
```

Lets explore temperature data for the past year from New York, Newark, Phildelphia, and Boston. The data comes from http://www.ncdc.noaa.gov/cdo-web/datasets and is stored in a CSV file that we will open using the pandas (http://pandas.pydata.org/) library. 

### Locations: 
* CITY:US360019 - New York, NY US
* CITY:US250002 - Boston, MA US
* CITY:US420015 - Philadelphia, PA US
* CITY:US340003 - Newark, NJ US

### Data Types: 	
* MNTM - Monthly mean temperature
* MMNT - Monthly Mean minimum temperature
* MMXT - Monthly Mean maximum temperature
* EMNT - Extreme minimum daily temperature
* EMXT - Extreme maximum daily temperature
* TPCP - Total precipitation
* EMXP - Extreme maximum daily precipitation
* TSNW - Total snow fall
* MXSD - Maximum snow depth
